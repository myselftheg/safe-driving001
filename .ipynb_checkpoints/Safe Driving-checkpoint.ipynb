### Univariate Screening

```python
from scipy.stats import chi2_contingency

cat_screen = pd.DataFrame()
for col in cat_columns + bin_columns:
#     print(col)
    t = pd.crosstab(X_train[col], y_train)
    # print(t)
    chi2, p, dof, expctd = chi2_contingency(t)
    print(col, 'chi:', '%.2f'%chi2, 'p-value:', '%.4f'%p)
    cat_dict = {'col':col, 'p-value':p}
    cat_screen = cat_screen.append(cat_dict, ignore_index=True)
```

```python
import category_encoders as ce

# class swoe(TransformerMixin):
#     def __init__(self):
#         pass
#     def fit(self, X, y):
#         self.X = X
#         self.y = y
#     def fit_transform(self, X, y):
#         res = pd.DataFrame()
#         for col in X.columns:
#             t = pd.Concat(X, y, axis=1)
#             res[col + '_swoe'] = np.log(t.)
    
class MyCatEncoders(TransformerMixin):
    def __init__(self, cat_columns, target_col):
        self.cat_columns, self.target_col = cat_columns, target_col
# Target encoding replaces a categorical value with the average value of the target for that value of the feature.         
        self.target_enc = ce.TargetEncoder(cols=cat_columns)
#Count encoding replaces each categorical value with the number of times it appears in the dataset.
        self.count_enc = ce.CountEncoder(cols=cat_columns)
#Finally, we'll look at CatBoost encoding. This is similar to target encoding in that it's based on the target probablity for a given value. However with CatBoost, for each row, the target probability is calculated only from the rows before it.
        self.catboost_enc = ce.CatBoostEncoder(cols=cat_columns)
        self.woe_enc = ce.WeightOfEvidence()
        pass
    
    def fit(self, X):
        self.X = X
        pass
    
    def transform(self, X):
        self.tar_encoded = self.target_enc.fit_transform(X[cat_columns], X[target_col])
        self.count_encoded = self.count_enc.fit_transform(X[cat_columns], X[target_col])
        self.catboost_encoded = self.catboost_enc.fit_transform(X[cat_columns], X[target_col])
        self.woe_encoded = self.woe_enc.fit_transform(X[cat_columns], X[target_col])
        
        res = pd.concat([self.tar_encoded.add_prefix('tenc_'), self.count_encoded.add_prefix('cenc_'),
                  self.catboost_encoded.add_prefix('catenc')], axis=1)
        return res
        
    
```
